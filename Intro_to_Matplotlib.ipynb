Matplotlib is a very important library that is used for visualozing the data. Using matplotlib you can create a variety of graphs this is especislly important for seeing how your models after training your model.

Importing Matplotlib

```python
import matplotlib.pyplot as plt
```

Creating and Custamizing your Graphs

```python
import numpy as np

xpoints  = np.array([0, 1, 2, 3, 4, 5, 6])
ypoints = np.array([0, 6, 12, 18, 24, 30, 36 ])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
ypoints = np.array([0, 6, 12, 18, 24, 30, 36 ])

plt.plot(ypoints)
plt.show()
```

Increase marker size and add marker edge color
marker size= ms
marker edge color = mec
marker face color = mfc

```python
plt.plot(ypoints, 'o:y',ms =10,mec='red',mfc ='blue')
plt.show()
```

Change the line width

```python
plt.plot(ypoints, 'o:y',ms =20,mec='red',mfc ='blue', linewidth = 5)
plt.show()
```

Plot multiple lines

```python
x1 = np.array([1, 2, 3, 4, 5, 6])
x2 = np.array([5, 10, 20, 40, 80, 160])
y1 = np.array([2, 4, 6, 8, 10, 12])
y2 = np.array([5, 10, 20, 40, 80, 160])

plt.plot(x1, y1)
plt.plot(x1, y2)
plt.show()

#Can also write in bottom way
plt.plot(x1, y1, x1, y2)
plt.show()
```

To Add labels and the titles use plt.xlabels and plt.ylabels and plt.title, like below

```python
plt.plot(x1, y1, marker = 'o')
plt.xlabel('Time(hours)')
plt.ylabel('Customer in shop')
plt.title('Number of customers in 1 hour time interval')
plt.show()
```
