Analysing Pandas Dataframes

```python
import pandas as pd
```

```python

data= "/content/sample_data/APDF - Sheet1.csv"
df = pd.read_csv(data)
df.head(5)
```

When creating an ML model, oftentimes you do not use every column of dataset as it is not feasible. Instead you choose the labels that have a large correlation to each other to find trends within the database. Panda makes it easy to find these corelation using the .corr method. Lets see how it works

```python
df.corr()
```

As you can see df.corr() shows the correlation of every column against each other. 1 Indicates a perfect positive corelation (Means they are same and increase/decrease together),-1 indicates a perfect negative corelation (meaning that one increases while the other decreases at the same rate) and the 0 indicates no corelation at all. In the correlation above we can see that Taxes and Sell have a strongg positive correlation (0.99) while Age and Sell has close ro no correlation

We can use the library called seaborn to better visualize these correlation with a heat map

```python
import seaborn as sns

#Load the auto dataset
auto_df = sns.load_dataset('mpg')

#Calculate the correlation matrix on the numeric column
corr = df.corr()

#Plot the heatmap
sns.heatmap(corr)
```

As you can see the lightest ones have the best correlation and the darkest ones have the best negative correlation, we can also use matplotlib to better visualize the relationship between two columns. we will go through more of matplotlib and pandas in the next lesson, but here is a quick introduction

```python
import matplotlib.pyplot as plt

df.plot()
plt.show()
```
